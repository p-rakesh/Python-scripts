```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Load the language model and tokenizer
model_name = "facebook/bart-large-cnn"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

def summarize(text, max_length=100, professional=True):
    # Create the prompt
    if professional:
        prompt = f"Please provide a concise and professional summary of the following text: '{text}'"
    else:
        prompt = f"Summarize the following text in {max_length} words: '{text}'"

    # Tokenize the prompt
    input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=1024, truncation=True)

    # Generate the summary
    summary_ids = model.generate(input_ids, max_length=max_length, min_length=10, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary

# Example usage
text = "I studied ABC in XYZ college with a percentage of y%. I am skilled in Q, W, E, and know the basics of R. I like data science field and want to work in a thriving organization.Desired Output:XYZ College was my academic home, where I achieved excellence in ABC, securing y% in my studies. Proficient in Q, W, and E, I also possess a fundamental grasp of R. The realm of data science has truly captured my imagination, and I'm on the lookout for a dynamic organization where I can channel my skills and enthusiasm."

professional_summary = summarize(text, max_length=50, professional=True)
casual_summary = summarize(text, max_length=50, professional=False)

print("Professional Summary:", professional_summary)
print("Casual Summary:", casual_summary)
```

```python
import transformers

def generate_professional_summary(text, prompt_type="generic", max_length=50):
    # Load the T5 model
    model_name = "t5-small"
    model = transformers.AutoModelForSeq2SeqLM.from_pretrained(model_name)
    tokenizer = transformers.AutoTokenizer.from_pretrained(model_name)

    # Define prompts based on prompt_type
    prompts = {
        "generic": f"Summarize the following text in a professional manner: '{text}'",
        "news": f"Craft a professional summary for the news article: '{text}'",
        "research": f"Generate a formal summary of the research paper: '{text}'",
        # Add more prompt variations as needed
    }

    # Select the appropriate prompt based on prompt_type
    prompt = prompts.get(prompt_type, prompts["generic"])

    # Tokenize and generate the professional summary
    input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(input_ids, max_length=max_length, min_length=10, length_penalty=2.0, num_beams=4, early_stopping=True)
    professional_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return professional_summary

# Example usage
text = "This is a long text that I want to summarize in a professional manner."

generic_summary = generate_professional_summary(text, prompt_type="generic")
news_summary = generate_professional_summary(text, prompt_type="news")
research_summary = generate_professional_summary(text, prompt_type="research")

print("Generic Summary:", generic_summary)
print("News Summary:", news_summary)
print("Research Summary:", research_summary)

```

```python
import transformers

def generate_summary_with_prompts(input_text, prompts, max_length=50):
    # Load the BART model and tokenizer
    model_name = "facebook/bart-large-cnn"
    model = transformers.AutoModelForSeq2SeqLM.from_pretrained(model_name)
    tokenizer = transformers.AutoTokenizer.from_pretrained(model_name)

    generated_summaries = []

    # Iterate through prompts and generate summaries
    for prompt in prompts:
        # Craft a prompt for generating a summary
        full_prompt = f"{prompt} '{input_text}'"

        # Tokenize and generate the summary
        input_ids = tokenizer.encode(full_prompt, return_tensors="pt", max_length=1024, truncation=True)
        summary_ids = model.generate(input_ids, max_length=max_length, min_length=10, length_penalty=2.0, num_beams=4, early_stopping=True)
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

        generated_summaries.append(summary)

    return generated_summaries

# Example usage
text = "In my educational journey at XYZ College, I specialized in ABC and graduated with a commendable y% performance. My abilities span across Q, W, and E, complemented by a foundational understanding of R. The expansive domain of data science appeals to my aspirations, and I am eager to contribute my talents to a thriving organization."

prompts = [
    "Reformulate the following passage: 'I studied ABC in XYZ college with a percentage of y%. I am skilled in Q, W, E, and know the basics of R. I like data science field and want to work in a thriving organization.",
    "Summarize the following text in a concise manner:I completed my academic pursuit at XYZ College, excelling in ABC with an impressive y% score. Proficient in Q, W, and E, I also possess a rudimentary understanding of R. The exciting realm of data science has captured my attention, and I am determined to find my place within a flourishing organization where I can apply my skills effectively.",
    "Craft a formal summary for the text:I completed my academic pursuit at XYZ College, excelling in ABC with an impressive y% score. Proficient in Q, W, and E, I also possess a rudimentary understanding of R. The exciting realm of data science has captured my attention, and I am determined to find my place within a flourishing organization where I can apply my skills effectively.",
]

generated_summaries = generate_summary_with_prompts(text, prompts, max_length=500)
for i, summary in enumerate(generated_summaries):
    print(f"Generated Summary {i+1}:", summary)
```
